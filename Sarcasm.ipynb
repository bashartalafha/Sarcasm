```python
import pandas as pd
import numpy as np
import logging
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from simpletransformers.classification import ClassificationModel, ClassificationArgs
```

```python
data = pd.read_excel("Sarcasm-data.xlsx")
```

```python
votes = data.columns[4:15]
```

```python
all_data = []
for index, row in data.iterrows():
    all_data.append([row['Tweet'], [row[vote].lower() for vote in votes].count("yes")/11])
```

```python
data_df = pd.DataFrame(all_data)
data_df.columns = ["text", "labels"]
```

```python
logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)

# Enabling regression
# Setting optional model configuration
model_args = ClassificationArgs()
model_args.num_train_epochs = 1
model_args.regression = True
model_args.overwrite_output_dir = True
```

```python
# Create a ClassificationModel
model = ClassificationModel(
    "bert",
    "bashar-talafha/multi-dialect-bert-base-arabic",
    num_labels=1,
    args=model_args
)
```

```python
kf10 = KFold(n_splits=10, shuffle=False)
```

```python
# Cross-validation - 10 folds
sum_loss = 0

for fold_idx, (train_index, test_index) in enumerate(kf10.split(data_df)):
 
    train_df = data_df.iloc[train_index]
    eval_df = data_df.iloc[test_index]

    # Train the model
    print(f"Training fold:{fold_idx+1}")
    model.train_model(train_df, overwrite_output_dir=True)
    
    # Evaluate the model
    result, model_outputs, wrong_predictions = model.eval_model(eval_df)  
    print(f"Eval loss for fold {fold_idx+1}: {result['eval_loss']}")
    sum_loss+=result['eval_loss']

print(f"\nFinal loss = {sum_loss/10}")
```

```python
print(f"\nFinal loss = {sum_loss/10}")
```

```python
# Make predictions with the model
predictions, raw_outputs = model.predict([""])
predictions
```
